```python
# Import dependecies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
plt.grid()
plt.show()
```

```python
# Create the plot and add a label for the legend (MATLAB Method)
plt.plot(x_axis,  y_axis, label = "Boston")
# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare ($)")
# Set the y limit between 0 and 45
plt.ylim(0, 45)
# Create a title
plt.title("Pyber Fare by Month")
# Add the legend
plt.legend()
```

```python
# Create the plot (MATLAB Method)
plt.plot(x_axis, y_axis, marker = "*", color = "blue", linewidth = 2, label = "Boston")
#Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45
plt.ylim(0, 45)
# Create a title
plt.title("Pyber Fare by Month")
# Add a grid
plt.grid()
# Add a legend
plt.legend()
```

```python
# Creaate a plot using Object-Oriented Method
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, color = "g", marker = "D", linewidth = 2, label = "Boston")

# Create title and axis labels
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
ax.set_title("Pyber Fare by Month")

# Add y-axis limit
ax.set_ylim(0, 45)

# Add grid
ax.grid()

# Add legend
ax.legend()
```

```python
# Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create a scatter plot
plt.plot(x_axis, y_axis, "o")
```

```python
# Create a scatter plot
plt.scatter(x_axis, y_axis)
```

```python
# Skill challenge
plt.scatter(y_axis, x_axis, color = "r", label = "Chicago")

# Add title and axes labels
plt.title("PyBer Fare by Month")
plt.xlabel("Fare ($)")
plt.ylabel("Date")

# Add limit for x-axis
plt.xlim(0, 50)

# Invert y-axis
plt.gca().invert_yaxis()

# Add legend
plt.legend()
```

```python
# Create a plot
plt.scatter(x_axis, y_axis, s = y_axis)
```

```python
# Adjust points in y axis
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s = y_axis_larger)
```

```python
# Format for list comprehension
# new_list = [expression for item in list if conditional]
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Scatter plot using object oriented method
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Scatter plot using object oriented method
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s = y_axis)
```

```python
# Scatter plot using obkect oriented method
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, color = "cyan", s = [i * 5 for i in y_axis], 
           alpha = 0.2, edgecolor = "black", linewidth = 2, label = "Boston")

# Add legend
ax.legend()

# Add title and axes labels
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")

# Set limit
ax.set_xlim(0, 50)

# Invert y_axis
plt.gca().invert_yaxis()
```

```python
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step = 5.0))
plt.gca().invert_yaxis()
```

```python
fig,  ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step = 5.0))
plt.gca().invert_yaxis()
```

```python
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure
fig, ax = plt.subplots(figsize = (8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step = 5.0))

# Create minor ticks at an increment of 1
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
